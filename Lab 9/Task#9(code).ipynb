```python
import pandas as pd
data = {'Student_ID': [101, 102, 103, 104],
    'City': ['Lahore', 'Karachi', 'Lahore', 'Islamabad'],
    'Performance': ['Good', 'Average', 'Excellent', 'Good']}
data_frame = pd.DataFrame(data)

print("Before One Hot Encoding:\n")
print(data_frame)

encoded_data = pd.get_dummies(data_frame, columns=['City', 'Performance'])

print("\nAfter One Hot Encoding:\n")
print(encoded_data)
```
